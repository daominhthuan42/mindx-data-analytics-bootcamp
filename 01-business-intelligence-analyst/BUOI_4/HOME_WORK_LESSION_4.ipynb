```python
import pandas as pd
import numpy as np
# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500) # To display all the columns of dataframe
pd.set_option('max_colwidth', None) # To set the width of the column to maximum
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Data_raw.csv")
df.head(5)
```

```python
df.info()
```

```python
# Remove khoảng trắng 2 đầu trên mỗi column
df.columns = df.columns.str.strip()
```

```python
# Đổi tên cột cho chuẩn
df.rename(columns={
    "Ví trí có thể phù hợp": "Vị trí phù hợp",
    "FALSE": "Đã liên hệ",  # hoặc xoá nếu không cần
    "Unnamed: 0": "ID",
}, inplace=True)
```

```python
df.info()
```

```python
# Loại bỏ các cột hoàn toàn rỗng (toàn NaN)
df.dropna(axis=1, how='all', inplace=True)
```

```python
df.info()
```

```python
# Remove 2 column Unnamed: 10 và Unnamed: 11
df.drop(columns=["Unnamed: 10", "Unnamed: 11"], inplace=True)
df.info()
```

```python
# Kiểm tra các cột quan trọng bị thiếu nhiều dữ liệu
missing_ratio = df.isnull().mean().sort_values(ascending=False)
print(missing_ratio)
```

```python
df.head()
```

```python
# Chuyển kiểu dữ liệu ngày tháng
df['Ngày tiếp nhận hồ sơ'] = pd.to_datetime(df['Ngày tiếp nhận hồ sơ'], errors='coerce')
df.head()
```

```python
# Remove hết khoảng trắng 2 đầu trong dữ liệu
for column in df.columns:
    if df[column].dtype == "object":
        df[column] = df[column].str.strip()
    else:
        pass
```
