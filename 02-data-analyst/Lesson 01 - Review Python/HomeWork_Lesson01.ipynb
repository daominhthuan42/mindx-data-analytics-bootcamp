```python
# Bài tập 1 :
# Bạn hãy xây dựng đoạn chương trình cho người dùng nhập số điểm GPA là số thập phân (với điều kiện >= 0 ) từ bàn phím. Nếu người dùng nhập số < 0, yêu cầu người dùng nhập lại.
# 	Sau khi người dùng nhập xong, chương trình tính toán và trả về kết quả đánh giá học
# 	Lực dựa trên GPA đó.
# Biết rằng:
# ➔Dưới 2 —- điểm F
# ➔2 đến 2.5 —- điểm C
# ➔2.5 đến 3 —- điểm B
# ➔3 đến 3.2 –– điểm B+
# ➔3.2 đến 3.7 —- điểm A
# Trên 3.7 — điểm A+
```

```python
def checking_gpa(gpa):
  if gpa < 0:
    raise ValueError("The GPA must be grater than or equal to zero.")
  elif gpa > 4:
    raise ValueError("The GPA must be less than 4.")
  return True
```

```python
def ranking_student():
  gpa = float(input("Enter GPA: "))
  if checking_gpa(gpa):
    if gpa < 2:
      print("Rank F")
    elif gpa >=2 and gpa < 2.5:
      print("Rank C")
    elif gpa >=2.5 and gpa < 3:
      print("Rank B")
    elif gpa >=3 and gpa < 3.2:
      print("Rank A")
    else:
      print("Rank A+")
n = 0
while(n < 3):
  ranking_student()
  n+=1
```

```python
# Bài tập 4:
# 	Tạo class có tên là Employee gồm các thuộc tính: EmID, Name, Address, DoB, Type(có 2 loại: Văn phòng, sản xuất), salary (lương căn bản) và các phương thức nhập, xuất thông tin của nhân viên đó.
# 	Sau đó, khởi tạo 5 đối tượng nhân viên có các thông tin như bảng bên dưới và lưu trữ vào một list có tên là Employee_Info:
```

```python
class Employee:
    def __init__(self, emp_id="", name="", address="", dob="", emp_type="", salary=0.0):
        """Constructor"""
        self.emp_id = emp_id
        self.name = name
        self.address = address
        self.dob = dob
        self.emp_type = emp_type
        self.salary = salary

    def input_info(self):
        """Enter employee information from keyboard"""
        self.emp_id = input("Enter employee ID: ")
        self.name = input("Enter employee name: ")
        self.address = input("Enter address: ")
        self.dob = input("Enter date of birth (dd-mm-yyyy): ")
        self.emp_type = input("Enter employee type (Office / Production): ")
        self.salary = float(input("Enter basic salary: "))

    def display_info(self):
        """Print employee information"""
        print("\n----- Employee Information -----")
        print(f"Employee ID   : {self.emp_id}")
        print(f"Name          : {self.name}")
        print(f"Address       : {self.address}")
        print(f"Date of Birth : {self.dob}")
        print(f"Type          : {self.emp_type}")
        print(f"Salary        : {self.salary}")

emp1 = Employee(emp_id="I001", name="Tú Anh", address="Vĩnh Phúc", dob="22/09/2000", emp_type="Văn Phòng", salary=500000)
emp2 = Employee(emp_id="I002", name="Cường", address="Hưng Yên", dob="23/06/2000", emp_type="Sản Xuất", salary=300000)
emp3 = Employee(emp_id="I003", name="Huyền", address="Hà Nội", dob="24/08/1999", emp_type="Văn Phòng", salary=570000)
emp4 = Employee(emp_id="I004", name="Hoa", address="Hà Tĩnh", dob="27/12/1998", emp_type="Văn Phòng", salary=700000)
emp5 = Employee(emp_id="I005", name="Phước", address="Nghệ An", dob="02/03/1995", emp_type="Sản Xuất", salary=800000)

Employee_Info = [emp1, emp2, emp3, emp4, emp5]

for emp in Employee_Info:
    emp.display_info()
```

```python
# Bài tập 5:
# 	Giả sử Công ty có hai loại nhân viên: Nhân viên văn phòng và Nhân viên sản xuất. Viết hàm:
# -Nhận vào thông tin là số ngày làm việc đối với nhân viên văn phòng
# -Nhận vào thông tin là số ngày làm việc và số sản phẩm đã làm được đối với nhân viên sản xuất.
# Sau đó tính lương cho từng nhân viên của công ty ở bảng bên trên và hiển thị ra màn hình kết quả có cấu trúc như sau:
# -{Mã_nhân_viên} lương tháng này là {số tiền lương đã tính}
# Biết rằng công ty cần tính lương cho nhân viên như sau:
# -Đối với nhân viên sản xuất:
# ▪Lương=lương căn bản * số ngày làm việc + số sản phẩm * 5.000
# -Đối nhân viên văn phòng:
# Lương = lương căn bản * số ngày làm việc + 2.000.000
```

```python
class Employee:
    def __init__(self, emp_id="", name="", emp_type="", base_salary=0.0):
        self.emp_id = emp_id
        self.name = name
        self.emp_type = emp_type  # "Office" or "Production"
        self.base_salary = base_salary

    def compute_salary(self, working_days, products=0):
        """Calculate salary based on employee type"""
        if self.emp_type == "Office":
            return self.base_salary * working_days + 2000000
        elif self.emp_type == "Production":
            return self.base_salary * working_days + products * 5000
        else:
            return 0.0


# Create 5 employees
emp1 = Employee("E001", "Nguyen Van A", "Office", 500000)
emp2 = Employee("E002", "Tran Thi B", "Production", 400000)
emp3 = Employee("E003", "Le Van C", "Office", 450000)
emp4 = Employee("E004", "Pham Thi D", "Production", 420000)
emp5 = Employee("E005", "Hoang Van E", "Office", 470000)

Employee_Info2 = [emp1, emp2, emp3, emp4, emp5]

working_data = {
    "E001": {"days": 22},                   # Office
    "E002": {"days": 25, "products": 100},  # Production
    "E003": {"days": 20},                   # Office
    "E004": {"days": 26, "products": 150},  # Production
    "E005": {"days": 21}                    # Office
}

# Calculate and print salary for each employee
for emp in Employee_Info2:
    data = working_data[emp.emp_id]
    salary = emp.compute_salary(data["days"], data.get("products", 0))
    print(f"{emp.emp_id} monthly salary is {salary:,} VND")
```
