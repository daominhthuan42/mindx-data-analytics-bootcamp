```python
# Câu 1: Bạn hãy xây dựng một hàm nhận tham số truyền vào là 1 chuỗi ký tự bất kỳ. 
# Hàm kiểm tra và trả về kết quả chuỗi trên là một đoạn password hợp lệ hay không, biết 
# rằng password hợp lệ cần thoả mãn các tiêu chí: 
# - Chứa 1 ký tự đặc biệt trong các ký tự !, @, #, $, ^ 
# - Chứa ít nhất một ký tự số 
# - Chứa ít nhất một ký tự chữ cái viết hoa 
# - Chứa ít nhất một ký tự chữ cái viết thường 
# - Độ dài từ 8 - 15 ký tự. (0.75đ)
import re

def isValidPassword(password):
    reLessUpperCaseChar = re.compile(r"[A-Z]")
    reLessLowerCaseChar = re.compile(r"[a-z]")
    reLessDigitChar = re.compile(r"\d")
    reLessSpecialChar = re.compile(r"[!@#$%^]")
    if (
            reLessUpperCaseChar.search(password) and
            reLessLowerCaseChar.search(password) and
            reLessDigitChar.search(password) and
            reLessSpecialChar.search(password) and
            (len(password) >= 8 and len(password) <= 15)
       ):
        print(f"Password: {password} hợp lệ")
    else:
        print(f"Password: {password} không hợp lệ")

isValidPassword("Hoilamgi1!")
isValidPassword("dadmadada312qdd")
isValidPassword("ddoaodj434!@F")
isValidPassword("ddoaodj434!@Fddadadddqwqellaldsadpadmmsdmadmasdmada")
```

```python
# Câu 3: Bạn hãy xây dựng một hàm nhận tham số truyền vào là một danh sách bất kỳ 
# chỉ chứa các số. Hàm tính toán và trả về số lớn nhì trong danh sách. Với trường hợp danh sách gồm các số bằng nhau, 
# hàm trả về số lớn nhất. Nếu có nhiều giá trị bằng số lớn nhì trong danh sách, hàm trả về 1 giá trị duy nhất. 
# Ví dụ:  
# - Với danh sách [1,1,1,1,1,1] -> Hàm trả về 1 
# - Với danh sách [1,2,1,1,2,3] -> Hàm trả về 2
def isvalidNumber(lst):
    flagCheck = True
    print(f"Danh sách: {lst}")
    reValidNUmber = re.compile(r"^-?\d+(\.\d+)?$")
    for item in lst:
        if reValidNUmber.search(str(item)):
            pass
        else:
            print(f"Item: {item} không phải số")
            flagCheck = False
    return flagCheck

def isSameValueInList(lst):
    flag = True
    for item in lst:
        if item == lst[0]:
            pass
        else:
            flag = False
            break
    return flag

def arrangeValueInList(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lst[j] < lst[j + 1]:
                # Hoán đổi nếu phần tử trước nhỏ hơn phần tử sau
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    # Trả về danh sách giảm dần
    return lst

def returnSecondLargetNumber(lst):
    if len(lst) == 0:
        print("Danh sách rỗng")
    else:        
        isvalid = isvalidNumber(lst)
        if isvalid is True:
            isSameValue = isSameValueInList(lst)
            if isSameValue is False:
                lst = arrangeValueInList(lst)
                print(f"Danh sách sau khi sắp xếp giảm dần: {lst}")
                print(f"Số lớn thứ 2 trong danh sách: {lst[1]}")
            else:
                print(f"Số lớn thứ 1 trong danh sách: {lst[0]}")
        else:
            print("[!]: Vui lòng nhập lại danh sách")

returnSecondLargetNumber([1, 5, 2, 14, 8.98, 321])
print("\n")
returnSecondLargetNumber([1, 5, 2, "abc", 8.98, 321])
print("\n")
returnSecondLargetNumber([1, 2, 1, 1, 2, 3])
print("\n")
returnSecondLargetNumber([1, 1, 1, 1, 1, 1])
```
