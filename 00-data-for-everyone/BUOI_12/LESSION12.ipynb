Check dữ liệu có bị duplicate hay không

```python
#Check dữ liệu bị trùng lặp
df.duplicated().sum()
# Check ra được có 818 hàng dữ liệu bị trùng lặp -->>> xóa bỏ các dòng trùng lặp
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df
```

```python
#Tách cột với split
df['place_with_parent_names'].str.split("|", expand = True)[1]
```

```python
#Thêm cột mới Country tách từ cột place_with_parent_names
df['Country'] = df['place_with_parent_names'].str.split("|", expand = True)[1]
```

```python
df
```

```python
#Thêm cột mới State tách từ cột place_with_parent_names
df['State'] = df['place_with_parent_names'].str.split("|", expand = True)[2]
df['City'] = df['place_with_parent_names'].str.split("|", expand = True)[3]
```

```python
df[['City','State']]
```

```python
df
```

```python
df = df.drop(columns = ["place_with_parent_names"])
```

```python
df
```

```python
df['lat-lon'].str.split("," , expand = True)
```

```python
df['lat'] = df['lat-lon'].str.split("," , expand = True)[0]
```

```python
df['lon'] = df['lat-lon'].str.split("," , expand = True)[1]
```

```python
df = df.drop(columns = ['lat-lon'])
```

```python
df.info()
```

```python
#Replace kí tự $ của cột price_usd thành '' và chuyển kiểu dữ liệu thành float
df['price_usd'] = df['price_usd'].str.replace("$", "").str.replace(",", "").astype(float)
```

```python
df
```

```python
df.info()
```
