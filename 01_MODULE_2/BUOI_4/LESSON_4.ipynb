```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
df_netflix = pd.read_csv("D:/02_DATA/02_MINDX/01_MODULE_2/BUOI_4/netflix_titles.csv")
```

```python
# Get the summary info of data set
def table_summary(df):
    print(f"Dataset shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes, columns=["dtypes"])
    summary = summary.reset_index()
    summary["Name"] = summary["index"]
    summary = summary[['Name','dtypes']]
    summary["Missing value"] = df.isnull().sum().values
    summary["Missing value %"] = 100 * np.round(df.isnull().sum().values / len(df), 2)
    summary["Duplicate value"] = df.duplicated().sum()
    summary["Duplicate value %"] = 100 * np.round(df.duplicated().sum() / len(df), 2)
    summary['Uniques'] = df.nunique().values
    return summary
```

```python
table_summary(df=df_netflix)
```

The column `show_id` is just number to identify this student in this dataset. So this column is not really a part of the information we should care about. We can drop this column or make it the index for this dataset.

```python
df_netflix.drop("show_id", axis=1, inplace=True)
```

```python
table_summary(df=df_netflix)
```

```python
# Replace all missing values (NaN) in the director column of df_student with the most common value (mode) from that column.
feature_missing = ["director", "cast", "country", "date_added", "rating", "duration"]
for feature in feature_missing:
    df_netflix[feature] = df_netflix[feature].fillna(df_netflix[feature].mode()[0])
```

```python
table_summary(df=df_netflix)
```
